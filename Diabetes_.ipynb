```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
mean_insulin = df[df['Insulin'] != 0]['Insulin'].mean()

df['Insulin'] = df['Insulin'].replace(0, mean_insulin)
```

```python
Features=df.drop(columns="Outcome" , axis=1)
Target=df["Outcome"]
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(Features)

```

```python
standardized_data=scaler.transform(Features)
```

```python
Features=standardized_data
Target=df["Outcome"]
```

```python
x_train,x_test,y_train, y_test=train_test_split(Features,Target,test_size=0.2,stratify=Target , random_state=39)
```

##SVM

```python
classifier = svm.SVC(kernel="rbf")
classifier.fit(x_train, y_train)
```

```python
x_train_predict=classifier.predict(x_train)
accuracy=accuracy_score(x_train_predict , y_train)
```

```python
print("our accurcy by using svm= :" , accuracy)
```

##random forest


```python
rf_model= RandomForestClassifier(n_estimators=20 , random_state=39)
rf_model.fit(x_train , y_train)
```

```python
rf_model_pred=rf_model.predict(x_test)
accuracy_rf=accuracy_score(rf_model_pred,y_test)
```

```python
print("accuracy of random forest:" , accuracy_rf)
```
